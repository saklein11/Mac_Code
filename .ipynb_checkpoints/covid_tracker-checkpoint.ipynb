```python
#This code returns grant information on all pmids gathered from pubmed
#It is modified slightly to retrieve additional info (article name, author affiliation)
#NOTE: There is an issue with the journal scraping as we aren't catching something

import collections
import requests
import urllib.request
from bs4 import BeautifulSoup
import time
import pickle

#NOTE: consider adding article date information to grant scraper for accurate dating down to the day
#      use 'pubdate' to grab date information for future grant searches

if appending == False: #only set these variables to empty lists if NOT appending to pickled lists
    fgrant = [[],[],[]] #0 = grant num, 1 = grant agency, 2 = grant country
    author = [[],[]] #0 = author name, 1 = affiliation
    article = [[],[],[],[]] #0 = pub date, 1 = journal, 2 = title, 3 = article type (letter, research art, rev, etc.)
ct = 0

for x in pmids:
    
    #Counter to track number of citations tackled
    ct += 1
    if ct % 300 == 0:
        print(ct/len(pmids),' completed!')
    
    skey = 'dcff69acdc62008c90b1354b0ac141dc0808' #api key to get more queries to ncbi eutils
    time.sleep(0.2) #setting to 0.1 throws over-query error
    url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id={0}&retmode=xml&api_key={1}".format(x,skey)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    
    '''#Should switch to an XML parser as html parser NOT optimal for this task!
    import xml.etree.ElementTree as ET
    tree = ET.fromstring(requests.get(url).text)
    #print(requests.get(url).text)
    junk = tree.findall('.//LastName')
    for item in junk:
        print(item.text)'''
    
    ###SCRAPE PUBLICATION ARTICLE INFORMATION###
    #Date
    date = soup.find('pubmedpubdate',attrs={'pubstatus':"accepted"})
    #print(response,'\n',date,'\n')
    if date == None: #Article information not listed in pubmed and need to skip this pmid
        date = soup.find('pubmedpubdate',attrs={'pubstatus':"pubmed"})
        if date == None:
            print('failed ',x)
            article[0].append('NA')
            article[1].append('NA')
            article[2].append('NA')
            article[3].append('NA')
            author[0].append('NA')
            author[1].append('NA')
            fgrant[0].append('NA')
            fgrant[1].append('NA')
            fgrant[2].append('NA')
            continue
    
    pubdate = ''
    for item in date:
        for subitem in item: #pubdate lists in order: year, month, day
            try:
                int(subitem)
                if pubdate.count('-') == 2:
                    break
                if pubdate != '':
                    pubdate = pubdate + '-' + str(subitem)
                if pubdate == '':
                    pubdate = str(subitem)
            except ValueError:
                continue
    article[0].append(pubdate)
    
    #Journal name
    try:
        article[1].append(str(soup.find_all('isoabbreviation')).split('>')[1].split('<')[0])
    except IndexError:
        article[1].append('NA')
    '''for jrn in soup.find_all('isoabbreviation'):
        article[1].append(str(jrn).split('>')[1].split('<')[0])'''
    if soup.find_all('isoabbreviation') == None:
        article[1].append('NA')
    #Article name/type
    article[2].append(str(soup.find_all('articletitle')).split('>')[1].split('<')[0])
    article[3].append(str(soup.find_all('publicationtype')).split('>')[1].split('<')[0])
    ###END ARTICLE BLOCK###
    
    ####SCRAPE AUTHOR INFORMATION###
    athr = soup.find_all('author')
    tmpa = [[],[]]
    for n in athr:
        
        for subn in n:
            if '<lastname>' in str(subn):
                tmpa[0].append(str(subn).split('>')[1].split('<')[0])
            if '<forename>' in str(subn):
                tmpa[0][-1] = tmpa[0][-1] + ' ' + str(subn).split('>')[1].split('<')[0]
            if '<affiliation>' in str(subn):
                tmpa[1].append(str(subn).split('<affiliation>')[1].split('</affiliation>')[0])
    if len(tmpa[0]) == 0:
        tmpa[0] = ['NA']
    if len(tmpa[1]) == 0:
        tmpa[1] = ['NA']
    author[0].append(tmpa[0])
    author[1].append(tmpa[1])
    ###END AUTHOR BLOCK ###
    
    ###SCRAPE GRANT INFORMATION###
    content = soup.find_all('grant')
    if len(content) == 0:
        fgrant[0].append('NA')
        fgrant[1].append('NA')
        fgrant[2].append('NA')
        continue

    for item in content:
        grant = [[],[],[]]
        gid = False
        agency = False
        country = False
            
        for j in item:
            if 'grantid' in str(j):
                grant[0].append(str(j).split('>')[1].split('<')[0])
                gid = True
                continue
            if 'agency' in str(j):
                grant[1].append(str(j).split('>')[1].split('<')[0])
                agency = True
                continue
            if 'country' in str(j):
                grant[2].append(str(j).split('>')[1].split('<')[0])
                country = True
                continue
            
        if gid == False:
            grant[0].append('NA')
        if agency == False:
            grant[1].append('NA')
        if country == False:
            grant[2].append('NA')
    fgrant[0].append(grant[0])
    fgrant[1].append(grant[1])
    fgrant[2].append(grant[2])
        ###END GRANT BLOCK###

###SAVE LISTS TO PICKLE IN CASE LOSE KERNEL###        
with open(path + "20200320_article_pickle.txt", "wb") as fp:   #Pickling
    pickle.dump(article, fp)
with open(path + "20200320_author_pickle.txt", "wb") as fp:   #Pickling
    pickle.dump(author, fp)
with open(path + "20200320_fgrant_pickle.txt", "wb") as fp:   #Pickling
    pickle.dump(fgrant, fp)
    
print('article lengths ',len(article[0]),len(article[1]),len(article[2]),len(article[3]))
print('author lengths ',len(author[0]),len(author[1]))
print('fgrant lengths ',len(fgrant[0]),len(fgrant[1]),len(fgrant[2]))
```

```python
#Build list of queries for fedreporter api
import re
from datetime import datetime

path = "C:\\Users\\sean.klein\\Documents\\Research\\LnkRsrchPol\\"

queries = [[],[]] #0 = query name, 1 = query pubdate

with open(path + 'provenge_authorlist.txt','r') as f:
    next(f) #skip first line as it is titles
    for line in f:
        alist = [re.sub('[^A-z| ]','',x).strip() for x in line.split('|')[2].split(',')] #create list of authors to search fedreporter
        
        for i in range(len(alist)):
            queries[0].append(alist[i].split(' ')[0] + ' ' + alist[i].split(' ')[-1])
            queries[1].append(datetime.strptime(line.split('|')[-4], '%Y-%m-%d')) #convert date to datetime for easy comparison
print(queries[0][0:10],queries[1][0:10])
```

```python
#Use queries list to find potential grants related to work

import time
import requests
import urllib.request
import json

grants = [[],[],[],[]] #0 = pi, 1 = funding ic, 2 = proj, 3 = title

#need 'contactPi' and 'ic' and 'projectEndDate' and 'projectNumber'
for i in range(len(queries[0])):
    time.sleep(0.4)
    url = "https://api.federalreporter.nih.gov/v1/Projects/search?query=piName:{0}&sortBy=fy&sortOrder=desc".format(queries[0][i])
    grantxt = json.loads(requests.get(url).text)
    #print(queries[0][i],len(grantxt['items']))
    
    for item in grantxt['items']:
        print(queries[0][i],item['contactPi'],item['projectNumber'],item['budgetEndDate'],'||',queries[1][i])
        #First, check if contact PI is the searched PI; reject if not
        if queries[0][i].split(' ')[0] not in item['contactPi'] or queries[0][i].split(' ')[1] not in item['contactPi']:
            continue
        #Second, check if grant end date was within a year of the pub date
        tdiff.days =  queries[1][i] - datetime.strptime(item['budgetEndDate'].split('T')[0],'%Y-%m-%d')
        tdiff2.days = queries[1][i] - datetime.strptime(item['budgetStartDate'].split('T')[0],'%Y-%m-%d')
        if tdiff > 365.25 or tdiff2 < 7:
            continue
        grants[0].append(item['contactPi'])
        grants[1].append(item['ic'])
        grants[2].append(item['projectNumber'])
        grants[3].append(item['title'])
```
